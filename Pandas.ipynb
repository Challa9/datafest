```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import random as rd
```

```python
data = pd.read_csv("california-history.csv")
bins = [4000, 10000,15000,20000, 25000, 30000 ,35000, 40000, 50000, 60000]
ybins = [0, 5, 10, 15, 20, 25, 30, 35, 40, 45, 50, 55, 60, 65, 70, 75, 80]
plt.figure(figsize=(15,10))
plt.hist(data.death, bins = bins)
plt.xticks(bins)
plt.yticks(ybins)
plt.xlabel("hospitalization")
plt.ylabel("Death increase")
plt.show()
```

```python
plt.plot(data.date, data.death)
plt.xticks(data.date[::80])

plt.show()
```
